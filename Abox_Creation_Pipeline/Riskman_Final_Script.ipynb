```python
import pandas as pd
```

```python
df_parent_HS = pd.read_csv('/content/Data for TM - Hazardous Situations.csv')
df_harms = pd.read_csv('/content/Data for TM - Harms.csv')
df_context = pd.read_csv('/content/Data for TM - Device Context.csv')
df_component = pd.read_csv('/content/Data for TM - Device Component.csv')
df_function = pd.read_csv('/content/Data for TM - Device Function.csv')
df_hazard = pd.read_csv('/content/Data for TM - Hazard.csv')
df_domain_specific_hazard = pd.read_csv('/content/Data for TM - Domain Specific Hazard.csv')
df_sda = pd.read_csv('/content/Data for TM - SDA.csv')
df_sub_sda = pd.read_csv('/content/Data for TM - Sub SDA.csv')
df_add_pro = pd.read_csv('/content/Data for TM - Additional Property.csv')
df_events = pd.read_csv('/content/Data for TM - Events.csv')
df_risk_level = pd.read_csv('/content/Data for TM - Risk levels.csv')
df_sources_of_HS = pd.read_csv('/content/Data for TM - Sources of HS.csv')
```

```python
def save_file(content, path):
  f = open(path, "w")
  for line in content:
    f.write(line)
  f.close()
```

```python
lib = [
"""@prefix ex: <http://example.org#> .
@prefix riskman: <https://w3id.org/riskman/ontology#> .
@prefix dc: <http://purl.org/dc/terms/> .
@prefix rdfs: <http://www.w3.org/2000/01/rdf-schema#> .
@prefix schema: <http://schema.org/> .
@prefix rlevel: <https://w3id.org/riskman/ontology#> .


"""]
```

```python
multiline_parent_string = []

def create_parent_hazardous_situation_file(df_parent_HS):

  for index, row in df_parent_HS.iterrows():
    parentHS = "ParentHazardousSituation_"+str(row['HS_ID'])
    identifier = "phs_"+str(row['HS_ID'])

    line = "ex:"+parentHS+" a riskman:HazardousSituation ;\n"
    multiline_parent_string.append(line)
    line = "    dc:identifier " + "\"" + identifier + "\"" + " ;\n"
    multiline_parent_string.append(line)
    line = "    rdfs:label " + "\"" + str(row['Parent_Hazardous_Situation']) + "\"" + " .\n\n"
    multiline_parent_string.append(line)

    save_file(multiline_parent_string, "/content/Files/Parent_Hazardous_Situation.ttl")
```

```python
multiline_harms_string = []

def create_harm_file(df_harms):

  for index, row in df_harms.iterrows():
    harm = "harm_"+str(row['ID'])
    identifier = "hr_"+str(row['ID'])

    line = "ex:"+harm+" a riskman:Harm ;\n"
    multiline_harms_string.append(line)
    line = "    dc:identifier " + "\"" + identifier + "\"" + " ;\n"
    multiline_harms_string.append(line)
    line = "    rdfs:label " + "\"" + str(row['Harms']) + "\"" + " .\n\n"
    multiline_harms_string.append(line)

    save_file(multiline_harms_string, "/content/Files/harms.ttl")
```

```python
multiline_context_string = []

def create_device_context_file(df_context):

  for index, row in df_context.iterrows():
    context = "deviceContext_"+str(row['DC_ID'])
    identifier = "dcx_"+str(row['DC_ID'])

    line = "ex:"+context+" a riskman:DeviceContext ;\n"
    multiline_context_string.append(line)
    line = "    dc:identifier " + "\"" + identifier + "\"" + " ;\n"
    multiline_context_string.append(line)
    line = "    rdfs:label " + "\"" + str(row['Name']) + "\"" + " .\n\n"
    multiline_context_string.append(line)

    save_file(multiline_context_string, "/content/Files/device_context.ttl")
```

```python
multiline_component_string = []

def create_device_component_file(df_component):

  for index, row in df_component.iterrows():
    component = "deviceComponent_"+str(row['ID'])
    identifier = "dcm_"+str(row['ID'])

    line = "ex:"+component+" a riskman:DeviceComponent ;\n"
    multiline_component_string.append(line)
    line = "    dc:identifier " + "\"" + identifier + "\"" + " ;\n"
    multiline_component_string.append(line)
    line = "    rdfs:label " + "\"" + str(row['Device_Component']) + "\"" + " .\n\n"
    multiline_component_string.append(line)

    save_file(multiline_component_string, "/content/Files/device_component.ttl")
```

```python
multiline_function_string = []

def create_device_function_file(df_function):

  for index, row in df_function.iterrows():
    function = "deviceFunction_"+str(row['ID'])
    identifier = "df_"+str(row['ID'])

    line = "ex:"+function+" a riskman:DeviceFunction ;\n"
    multiline_function_string.append(line)
    line = "    dc:identifier " + "\"" + identifier + "\"" + " ;\n"
    multiline_function_string.append(line)
    line = "    rdfs:label " + "\"" + str(row['Device_Function']) + "\"" + " .\n\n"
    multiline_function_string.append(line)

    save_file(multiline_function_string, "/content/Files/device_function.ttl")
```

```python
multiline_hazard_string = []

def create_hazard_file(df_hazard):

  for index, row in df_hazard.iterrows():
    hazard = "hazard_"+str(row['ID'])
    identifier = "hz_"+str(row['ID'])

    line = "ex:"+hazard+" a riskman:Hazard ;\n"
    multiline_hazard_string.append(line)
    line = "    dc:identifier " + "\"" + identifier + "\"" + " ;\n"
    multiline_hazard_string.append(line)
    line = "    rdfs:label " + "\"" + str(row['Hazard']) + "\"" + " .\n\n"
    multiline_hazard_string.append(line)

    save_file(multiline_hazard_string, "/content/Files/hazard.ttl")
```

```python
multiline_domain_string = []

def create_domain_specific_hazard_file(df_domain_specific_hazard):

  for index, row in df_domain_specific_hazard.iterrows():
    dsHazard = "domainSpecificHazard_"+str(row['DSH_ID'])
    identifier = "dsh_"+str(row['DSH_ID'])

    line = "ex:"+dsHazard+" a riskman:DomainSpecificHazard ;\n"
    multiline_domain_string.append(line)
    line = "    dc:identifier " + "\"" + identifier + "\"" + " ;\n"
    multiline_domain_string.append(line)
    line = "    riskman:hasDeviceComponent ex:deviceComponent"+str(row['Device_Component'])+" ;\n"
    multiline_domain_string.append(line)
    line = "    riskman:hasDeviceFunction ex:deviceFunction"+str(row['Device_Function'])+" ;\n"
    multiline_domain_string.append(line)
    line = "    riskman:hasHazard ex:hazard"+str(row['Hazard'])+" .\n\n"
    multiline_domain_string.append(line)

    save_file(multiline_domain_string, "/content/Files/domain_specific_hazard.ttl")
```

```python
multiline_sda_string = []

def create_sda_file(df_sda, df_sub_sda):

  for index, row in df_sda.iterrows():
    sda = "sda_"+str(row['SDA_ID'])
    identifier = "sda_"+str(row['SDA_ID'])

    line = "ex:"+sda+" a riskman:SDA ;\n"
    multiline_sda_string.append(line)
    line = "    dc:identifier " + "\"" + identifier + "\"" + " ;\n"
    multiline_sda_string.append(line)
    line = "    rdfs:label " + "\"" + str(row['SDA']) + "\"" + " ;\n"
    multiline_sda_string.append(line)

    if not pd.isna(row['Imp_Manifest']):

      if not pd.isna(row['Assurance_SDA']):
        line = "    riskman:hasImplementationManifest [ \n        rdfs:label \"" + str(row['Imp_Manifest']) + "\"\n    ];\n"
        multiline_sda_string.append(line)
        line = "    riskman:hasSafetyAssurance [ \n        rdfs:label \"" + str(row['Assurance_SDA']) + "\"\n    ].\n\n"
        multiline_sda_string.append(line)
      else:
        line = "    riskman:hasImplementationManifest [ \n        rdfs:label \"" + str(row['Imp_Manifest']) + "\"\n    ].\n\n"
        multiline_sda_string.append(line)

    else:
      result = df_sub_sda[df_sub_sda['SDA_ID'] == row['SDA_ID']]
      count = 1

      for index1, row1 in result.iterrows():
        if len(result) == count:
          line = "    riskman:hasSubSDA ex:subSDA_"+str(row1['ID'])+" .\n\n"
          multiline_sda_string.append(line)
        else:
          line = "    riskman:hasSubSDA ex:subSDA_"+str(row1['ID'])+" ;\n"
          multiline_sda_string.append(line)
          count+=1

      for index2, row2 in result.iterrows():
        sub_sda = "subSDA_"+str(row2['ID'])
        sub_sda_identifier = "subSDA_"+str(row2['ID'])

        line = "ex:"+sub_sda+" a riskman:SubSDA ;\n"
        multiline_sda_string.append(line)
        line = "    dc:identifier " + "\"" + sub_sda_identifier + "\"" + " ;\n"
        multiline_sda_string.append(line)
        line = "    rdfs:label " + "\"" + str(row2['Sub_SDA']) + "\"" + " ;\n"
        multiline_sda_string.append(line)

        result1 = df_add_pro[df_add_pro['Sub_SDA_ID'] == row2['ID']]
        count = 1

        if not result1.empty:
          line = "    riskman:hasImplementationManifest [ \n        rdfs:label \"" + str(row2['Imp_Manifest']) + "\";\n"
          multiline_sda_string.append(line)

          for index3, row3 in result1.iterrows():
            if count == len(result1):
              line = "        schema:additionalProperty [ \n            schema:name \"" + str(row3['Name']) + "\" ;\n            schema:value " + str(row3['Value']) + " ;\n        ]\n"
              multiline_sda_string.append(line)
            else:
              line = "        schema:additionalProperty [ \n            schema:name \"" + str(row3['Name']) + "\" ;\n            schema:value " + str(row3['Value']) + " ;\n        ];\n"
              multiline_sda_string.append(line)
              count+=1

          line = "    ].\n\n"
          multiline_sda_string.append(line)

        else:
          line = "    riskman:hasImplementationManifest [ \n        rdfs:label \"" + str(row2['Imp_Manifest']) + "\"\n    ].\n\n"
          multiline_sda_string.append(line)

    save_file(multiline_sda_string, "/content/Files/sda.ttl")
```

```python
multiline_events_string = []
multiline_primary_cause_events_string = []
multiline_hazardous_sit_string = []
multiline_analysed_risk_string = []
multiline_controlled_risk_string = []
multiline_initial_risk_string = []
multiline_residual_risk_string = []

is_initial_risk = []
is_residual_risk = []

def create_controlled_analysed_residual_initial_risk_file(df_sources_of_HS, df_events, df_harms):
  events_count = 1
  analysed_risk_count = 1

  for index, row in df_sources_of_HS.iterrows():

      result = df_events[df_events['E_ID'] == row['E_ID']]
      result1 = df_parent_HS[df_parent_HS['ID'] == row['Hazardous_Situation']]

      if not result.empty:
        for index1, row1 in result.iterrows():
          if row1['is_Event_Given'] != 0:

            event = "event_"+str(events_count)
            identifier = "ev_"+str(events_count)

            line = "ex:"+event+" a riskman:Event ;\n"
            multiline_events_string.append(line)
            line = "    dc:identifier " + "\"" + identifier + "\"" + " ;\n"
            multiline_events_string.append(line)
            line = "    rdfs:label " + "\"" + str(row1['Events']) + "\"" + " .\n\n"
            multiline_events_string.append(line)


            pc_event = "primary_cause_event_"+str(events_count)
            pc_identifier = "pc_ev_"+str(events_count)

            line = "ex:"+pc_event+" a riskman:Event ;\n"
            multiline_primary_cause_events_string.append(line)
            line = "    dc:identifier " + "\"" + pc_identifier + "\"" + " ;\n"
            multiline_primary_cause_events_string.append(line)
            line = "    rdfs:label " + "\"" + str(row['Primary_Cause']) + "\"" + " ;\n"
            multiline_primary_cause_events_string.append(line)
            line = "    riskman:hasPrecedingEvent " + "ex:"+event+" .\n\n"
            multiline_primary_cause_events_string.append(line)


            hs = "hazardousSituation_"+str(events_count)
            hs_identifier = "hs_"+str(events_count)

            line = "ex:"+hs+" a riskman:HazardousSituation ;\n"
            multiline_hazardous_sit_string.append(line)
            line = "    dc:identifier " + "\"" + hs_identifier + "\"" + " ;\n"
            multiline_hazardous_sit_string.append(line)
            line = "    rdfs:label " + "\"" + result1.iloc[0]['Hazardous_Situation'] + "\"" + " ;\n"
            multiline_hazardous_sit_string.append(line)
            line = "    riskman:hasParentSituation " + "ex:ParentHazardousSituation"+str(result1.iloc[0]['HS_ID'])+" ;\n"
            multiline_hazardous_sit_string.append(line)
            line = "    riskman:hasEvent " + "ex:"+pc_event+" .\n\n"
            multiline_hazardous_sit_string.append(line)

            events_count += 1

          else:
            pc_event = "primary_cause_event_"+str(events_count)
            pc_identifier = "pc_ev_"+str(events_count)

            line = "ex:"+pc_event+" a riskman:Event ;\n"
            multiline_primary_cause_events_string.append(line)
            line = "    dc:identifier " + "\"" + pc_identifier + "\"" + " ;\n"
            multiline_primary_cause_events_string.append(line)
            line = "    rdfs:label " + "\"" + str(row['Primary_Cause']) + "\"" + " .\n\n"
            multiline_primary_cause_events_string.append(line)


            hs = "hazardousSituation_"+str(events_count)
            hs_identifier = "hs_"+str(events_count)

            line = "ex:"+hs+" a riskman:HazardousSituation ;\n"
            multiline_hazardous_sit_string.append(line)
            line = "    dc:identifier " + "\"" + hs_identifier + "\"" + " ;\n"
            multiline_hazardous_sit_string.append(line)
            line = "    rdfs:label " + "\"" + result1.iloc[0]['Hazardous_Situation'] + "\"" + " ;\n"
            multiline_hazardous_sit_string.append(line)
            line = "    riskman:hasParentSituation " + "ex:ParentHazardousSituation"+str(result1.iloc[0]['HS_ID'])+" ;\n"
            multiline_hazardous_sit_string.append(line)
            line = "    riskman:hasPrecedingEvent " + "ex:"+pc_event+" .\n\n"
            multiline_hazardous_sit_string.append(line)

            events_count += 1

          result2 = df_harms[df_harms['HS_ID'] == result1.iloc[0]['HS_ID']]

          for index2, row2 in result2.iterrows():
            i_result = df_risk_level[(df_risk_level['Probability_1'] == row1['Initial_Probability_1']) &
                                                    (df_risk_level['Probability_2'] == row2['Probability_2']) &
                                                    (df_risk_level['Severity'] == row2['Severity'])]

            if not i_result.empty and i_result.iloc[0]['ID'] not in is_initial_risk:
              initial = "initialRiskLevel_"+str(i_result.iloc[0]['ID'])
              identifier = "irl_"+str(i_result.iloc[0]['ID'])

              line = "ex:"+initial+" a riskman:RiskLevel ;\n"
              multiline_initial_risk_string.append(line)
              line = "    dc:identifier " + "\"" + identifier + "\"" + " ;\n"
              multiline_initial_risk_string.append(line)
              line = "    riskman:hasProbability1 rlevel:p"+ str(row1['Initial_Probability_1']) +" ;\n"
              multiline_initial_risk_string.append(line)
              line = "    riskman:hasProbability2 rlevel:p"+ str(row2['Probability_2']) +" ;\n"
              multiline_initial_risk_string.append(line)
              line = "    riskman:hasSeverity rlevel:s"+ str(row2['Severity']) +" .\n\n"
              multiline_initial_risk_string.append(line)
              is_initial_risk.append(i_result.iloc[0]['ID'])


            r_result = df_risk_level[(df_risk_level['Probability_1'] == row1['Residual_Probability_1']) &
                                                    (df_risk_level['Probability_2'] == row2['Probability_2']) &
                                                    (df_risk_level['Severity'] == row2['Severity'])]

            if not i_result.empty and r_result.iloc[0]['ID'] not in is_residual_risk:
              residual = "residualRiskLevel_"+str(r_result.iloc[0]['ID'])
              identifier = "rrl_"+str(r_result.iloc[0]['ID'])

              line = "ex:"+residual+" a riskman:RiskLevel ;\n"
              multiline_residual_risk_string.append(line)
              line = "    dc:identifier " + "\"" + identifier + "\"" + " ;\n"
              multiline_residual_risk_string.append(line)
              line = "    riskman:hasProbability1 rlevel:p"+ str(row1['Residual_Probability_1']) +" ;\n"
              multiline_residual_risk_string.append(line)
              line = "    riskman:hasProbability2 rlevel:p"+ str(row2['Probability_2']) +" ;\n"
              multiline_residual_risk_string.append(line)
              line = "    riskman:hasSeverity rlevel:s"+ str(row2['Severity']) +" .\n\n"
              multiline_residual_risk_string.append(line)
              is_residual_risk.append(r_result.iloc[0]['ID'])


            controlled_risk = "controlledRisk_"+str(analysed_risk_count)
            controlled_risk_identifier = "cr_"+str(analysed_risk_count)

            line = "ex:"+controlled_risk+" a riskman:ControlledRisk ;\n"
            multiline_controlled_risk_string.append(line)
            line = "    dc:identifier " + "\"" + controlled_risk_identifier + "\"" + " ;\n"
            multiline_controlled_risk_string.append(line)
            line = "    riskman:hasAnalyzedRisk ex:analyzedRisk_"+str(analysed_risk_count)+" ;\n"
            multiline_controlled_risk_string.append(line)
            line = "    riskman:hasResidualRiskLevel ex:residualRiskLevel_"+str(r_result.iloc[0]['ID'])+" ;\n"
            multiline_controlled_risk_string.append(line)
            line = "    riskman:isMitigatedBy ex:sda_"+str(row1['SDA_ID'])+" .\n\n"
            multiline_controlled_risk_string.append(line)


            analyzed_risk = "analyzedRisk_"+str(analysed_risk_count)
            analyzed_risk_identifier = "ar_"+str(analysed_risk_count)

            line = "ex:"+analyzed_risk+" a riskman:AnalyzedRisk ;\n"
            multiline_analysed_risk_string.append(line)
            line = "    dc:identifier " + "\"" + analyzed_risk_identifier + "\"" + " ;\n"
            multiline_analysed_risk_string.append(line)
            line = "    riskman:hasDeviceContext ex:deviceContext_"+str(row1['Device Context'])+" ;\n"
            multiline_analysed_risk_string.append(line)
            line = "    riskman:hasDomainSpecificHazard ex:domainSpecificHazard_"+str(row1['DSH_ID'])+" ;\n"
            multiline_analysed_risk_string.append(line)
            line = "    riskman:hasHarm ex:harm_"+str(row2['ID'])+" ;\n"
            multiline_analysed_risk_string.append(line)
            line = "    riskman:hasInitialRiskLevel ex:initialRiskLevel_"+str(i_result.iloc[0]['ID'])+" ;\n"
            multiline_analysed_risk_string.append(line)
            line = "    riskman:hasHazardousSituation ex:hazardousSituation_"+str(events_count)+" .\n\n"
            multiline_analysed_risk_string.append(line)

            analysed_risk_count += 1
```

```python
create_parent_hazardous_situation_file(df_parent_HS)
create_harm_file(df_harms)
create_device_context_file(df_context)
create_device_component_file(df_component)
create_device_function_file(df_function)
create_hazard_file(df_hazard)
create_domain_specific_hazard_file(df_domain_specific_hazard)
create_sda_file(df_sda, df_sub_sda)
create_controlled_analysed_residual_initial_risk_file(df_sources_of_HS, df_events, df_harms)
```

```python
merged = lib + multiline_controlled_risk_string + multiline_analysed_risk_string + multiline_domain_string + multiline_context_string + multiline_component_string + multiline_function_string + multiline_hazard_string + multiline_parent_string + multiline_hazardous_sit_string + multiline_primary_cause_events_string + multiline_events_string + multiline_harms_string + multiline_initial_risk_string + multiline_residual_risk_string + multiline_sda_string
```

```python
save_file(merged, "/content/merged.ttl")
```
